# Example 3 - Advanced REST API Call

This is an example of using a direct Rest API call to the model.  This will use more parameters and will allow you more freedom in terms of how you configure each message.  This is an advanced method but can provide more capabilites and be better implemented within a larger program for advanced users and coders in some respects.

For more information you can refer to the documentation here: https://learn.microsoft.com/en-us/azure/ai-foundry/openai/reference-preview?view=foundry-classic#chat-completions---create



```python
import requests
import time

# Configuration
API_KEY = "<YOUR_API_KEY>"
API_URL = "<YOUR_API_ENDPOINT>openai/deployments/gpt-4.1/chat/completions?api-version=2025-04-01-preview"

HEADERS = {
    "Content-Type": "application/json",
    "api-key": API_KEY
}

def send_message(messages, temperature=0.7, max_tokens=150):
    """
    Send a chat message to GPT-5-mini via the Responses API.
    :param messages: List of dicts representing the chat history.
    :return: Model's response as string.
    """
    payload = {
        "model": "gpt-4.1",
        "messages": messages,
        "temperature": temperature,
        "max_tokens": max_tokens
    }
    response = requests.post(API_URL, json=payload, headers=HEADERS)
    if response.status_code == 200:
        result = response.json()
        # Typical OpenAI-style API response:
        # { "choices": [{"message": {"role": "assistant", "content": "..."} }] }
        return result["choices"][0]["message"]["content"]
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None

def main():
    print(" GPT-4-mini Test Conversation\n")
    # Start with a greeting message
    messages = [
        {"role": "system", "content": "You are an assistant that helps with creative coding."},
        {"role": "user", "content": "Hello! Who are you?"}
    ]
    for i in range(3):
        response = send_message(messages)
        print(f"Assistant: {response}\n")
        # For the demo, user reply is generic. You can replace this with your custom input logic.
        user_reply = [
            "What can you do for me?",
            "Show me a simple Python example.",
            "Thanks! Bye."
        ]
        if i < len(user_reply):
            print(f"User: {user_reply[i]}\n")
            messages.append({"role": "assistant", "content": response})
            messages.append({"role": "user", "content": user_reply[i]})
        time.sleep(1)

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env -S poetry run python

from openai import OpenAI
from openai import AzureOpenAI

client = AzureOpenAI(
    api_key="<YOUR_API_KEY>",
    base_url="<YOUR_API_ENDPOINT>",
    api_version = "2025-04-01-preview",
)

# gets API Key from environment variable OPENAI_API_KEY
#client = OpenAI()

# Non-streaming:
print("----- standard request -----")
completion = client.chat.completions.create(
    model="gpt-4.1",
    messages=[
        {
            "role": "user",
            "content": "How do I output all files in a directory using Python?",
        },
    ],
)
print(completion.choices[0].message.content)

# Streaming:
print("----- streaming request -----")

stream = client.chat.completions.create(
    model="gpt-4.1",
    messages=[
        {
            "role": "user",
            "content": "How do I output all files in a directory using Python?",
        },
    ],
    stream=True,
)
for chunk in stream:
    if not chunk.choices:
        continue

    print(chunk.choices[0].delta.content, end="")
print()

# Response headers:
print("----- custom response headers test -----")
response = client.chat.completions.with_raw_response.create(
    model="gpt-4.1",
    messages=[
        {
            "role": "user",
            "content": "Say this is a test",
        }
    ],
)
completion = response.parse()
print(response.request_id)
print(completion.choices[0].message.content)
```
